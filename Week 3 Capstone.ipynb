# Custer Neighborhoods

Create a new data frame with boroughs containing Toronto

```python
toronto_data= newdf[newdf['Borough'].str.contains('Toronto', na = False)].reset_index(drop=True)
toronto_data.head()
```

Latitud and longitud for Toronto, as follows: 43.6532° N, 79.3832° W

```python
latitude = 43.6532
longitude= -79.3832
```

```python
import folium
# create map of TORONTO using latitude and longitude values above:
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
